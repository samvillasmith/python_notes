```python
## Handling missing values
df.isnull().any(axis=1)
```

```python
df_filled = df.fillna(0)
```

```python
# Filling missing values with the mean of the column
df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df
```

```python
df.dtypes
```

```python
# Renaming columns
df = df.rename(columns={'Date':'Sales Date'})
```

```python
df.head()
```

```python
# change data types
df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New Value'] = df['Value'].apply(lambda x:x*2)
df.head()
```

```python
# Data aggregating and grouping
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum = df.groupby(['Product', 'Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_mean2 = df.groupby(['Product', 'Region'])['Value'].mean()
print(grouped_mean2)
```

```python
# Aggregate multiple functions
grouped_agg = df.groupby('Region')['Value'].agg(['mean', 'sum', 'count'])
print(grouped_agg)
```

```python
# Merging and joining data frames
df1 = pd.DataFrame({'Product': ['A', 'B', 'C'], 'Price': [10, 20, 30]})
df2 = pd.DataFrame({'Product': ['A', 'B', 'D'], 'Stock': [100, 200, 300]})
```

```python
df1
```

```python
df2
```

```python
# Merge dataframes on the Product columns
pd.merge(df1, df2, on="Product", how="inner")
```

```python
pd.merge(df1, df2, on="Product", how="outer")
```

```python
pd.merge(df1, df2, on="Product", how="left")
```

```python
pd.merge(df1, df2, on="Product", how="right")
```

### Reading Data from Different Formats

```python
import pandas as pd
import json
from io import StringIO

Data = {"employee_name": "James", "email": "james@example.com", "job_profile": [{"job_title": "Software Engineer", "job_level": "Mid"}]}

# Convert dictionary to proper JSON string
json_string = json.dumps(Data)

# Use StringIO with the proper JSON string
string_io = StringIO(json_string)

# Now you can use pd.read_json with StringIO
df = pd.read_json(string_io)
```

```python
df
```

```python
df.to_json(orient="index")
```

```python
df.to_json(orient="records")
```
